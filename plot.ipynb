### Electric Field of Exp 1 [x,y,V]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Load your data
data = np.genfromtxt('data1-raw.csv', delimiter=',')  # replace 'yourfile.csv' with your file name    
# Remove rows with NaN values
data = data[~np.isnan(data).any(axis=1)]


# Assuming your data columns are [x, y, V]
x = data[:, 0]
y = data[:, 1]
V = data[:, 2].

# Define a grid for the field
xi = np.linspace(x.min(), x.max(), 10)  # adjust as needed
yi = np.linspace(y.min(), y.max(), 14)  # adjust as needed
X, Y = np.meshgrid(xi, yi)

# Interpolate V onto the grid
Vi = griddata((x, y), V, (X, Y), method='cubic')

# Calculate the gradient of V
Ey, Ex = np.gradient(-Vi)  # note the negative sign, because E = -grad(V)

# Normalize the electric field vectors
E = np.sqrt(Ex**2 + Ey**2)
Ex /= E
Ey /= E

# Plot the field lines
plt.figure(figsize=(8, 8))
plt.quiver(X, Y, Ex, Ey, E, pivot='middle', headlength=4, headwidth=2, scale=30, cmap='jet')
plt.colorbar(label='Electric field strength (V/cm)')
plt.title('Experiment 1: Electric Field Lines')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.grid(True)
# Save the figure
plt.savefig('1-EField.png')  # replace 'figure.png' with your desired file name
plt.show()
      
```

### Equipotential Field of Exp 1 [x,y,V]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Load your data
data = np.genfromtxt('data1-raw.csv', delimiter=',')  # replace 'yourfile.csv' with your file name

# Remove rows with NaN values
data = data[~np.isnan(data).any(axis=1)]

# Assuming your data columns are [x, y, V]    
x = data[:, 0]
y = data[:, 1]
V = data[:, 2]

# Define a grid for the field
xi = np.linspace(x.min(), x.max(), 10)  # adjust as needed
yi = np.linspace(y.min(), y.max(), 14)  # adjust as needed
X, Y = np.meshgrid(xi, yi)

# Interpolate V onto the grid
Vi = griddata((x, y), V, (X, Y), method='cubic')

# Plot the equipotential lines
plt.figure(figsize=(8, 8))
plt.contour(X, Y, Vi, levels=50, cmap='jet')  # adjust the number of levels as needed
plt.colorbar(label='Electric Potential (V)')
plt.title('Experiment 1: Equipotential Lines')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.grid(True)
# Save the figure
plt.savefig('1-EPotential.png')  # replace 'figure.png' with your desired file name
plt.show()
```

### Electric and Equipotential of Exp 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Load your data
data = np.genfromtxt('data1-raw.csv', delimiter=',')  # replace 'yourfile.csv' with your file name

# Remove rows with NaN values
data = data[~np.isnan(data).any(axis=1)]

# Assuming your data columns are [x, y, V]
x = data[:, 0]
y = data[:, 1]
V = data[:, 2]

# Define a grid for the field
xi = np.linspace(x.min(), x.max(), 10)  # adjust as needed
yi = np.linspace(y.min(), y.max(), 14)  # adjust as needed
X, Y = np.meshgrid(xi, yi)

# Interpolate V onto the grid
Vi = griddata((x, y), V, (X, Y), method='cubic')

# Calculate the gradient of V
Ey, Ex = np.gradient(-Vi)  # note the negative sign, because E = -grad(V)

# Normalize the electric field vectors
E = np.sqrt(Ex**2 + Ey**2)
Ex /= E
Ey /= E

# Plot the equipotential lines
plt.figure(figsize=(8, 8))
plt.contour(X, Y, Vi, levels=50, cmap='jet')  # adjust the number of levels as needed

# Overlay the electric field lines
plt.quiver(X, Y, Ex, Ey, E, pivot='middle', headlength=4, headwidth=2, scale=30, cmap='jet')

plt.colorbar(label='Electric field strength (V/cm) / Potential (V)')
plt.title('Experiment 1: Electric Field Lines and Equipotential Lines')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.grid(True)
# Save the figure
plt.savefig('1-EFieldandEPotential.png')  # replace 'figure.png' with your desired file name
plt.show()
```
